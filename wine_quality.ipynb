# Preprocessing

```python
dataset.head()
```

```python
X = dataset.drop('quality',axis=1).values
y = dataset['quality'].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix

cla = LogisticRegression(max_iter=200)
cla.fit(X_train, y_train)
pred = cla.predict(X_test)

print(accuracy_score(pred,y_test))
print(confusion_matrix(y_test, pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix

cla = KNeighborsClassifier(n_neighbors=8)
cla.fit(X_train,y_train)
pred = cla.predict(X_test)

print(accuracy_score(pred,y_test))
print(confusion_matrix(y_test, pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix

cla = DecisionTreeClassifier()
cla.fit(X_train,y_train)
pred = cla.predict(X_test)

print(accuracy_score(pred,y_test))
print(confusion_matrix(y_test, pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix

cla = RandomForestClassifier()
cla.fit(X_train,y_train)
pred = cla.predict(X_test)

print(accuracy_score(pred,y_test))
print(confusion_matrix(y_test, pred))
```

# SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix

cla = SVC(kernel='linear')
cla.fit(X_train,y_train)
pred = cla.predict(X_test)

print(accuracy_score(pred,y_test))
print(confusion_matrix(y_test, pred))
```
